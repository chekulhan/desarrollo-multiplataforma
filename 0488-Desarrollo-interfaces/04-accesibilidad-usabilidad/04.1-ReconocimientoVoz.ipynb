**Analisis Sentimental**

El análisis sentimental es una técnica del procesamiento del lenguaje natural (NLP) que permite identificar la emoción detrás de un texto. Puede clasificar opiniones como positivas, negativas o neutras. En esta introducción, usaremos la librería Hugging Face Transformers y su herramienta de pipelines, que simplifica el uso de modelos de inteligencia artificial para tareas como el análisis de sentimientos.


```python
from transformers import pipeline
```

```python
from transformers import pipeline

# Cargamos un pipeline de análisis de sentimientos
sentiment_pipeline = pipeline("sentiment-analysis")
#sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Probamos con algunos textos
textos = [
    "I love this product, it's amazing!",
    "I didn't like it at all, it was a bad experience.",
    "It's an acceptable product, but it could be better."
]

#textos = [
#    "¡Me encanta este producto, es increíble!",
#    "No me gustó para nada, fue una mala experiencia.",
#    "Es un producto aceptable, pero podría ser mejor."
#]

# Aplicamos el análisis de sentimientos
resultados = sentiment_pipeline(textos)

# Mostramos los resultados
for i in range(len(textos)):
    print(f"Texto: {textos[i]}")
    print(f"Sentimiento: {resultados[i]['label']} (Confianza: {resultados[i]['score']:.2f})")
    print("-" * 50)
```

Convertir audio en texto con reconocimiento de voz.


```python
from datasets import load_dataset, Audio
from IPython.display import Audio  # For Jupyter notebooks to play audio

# Load the dataset (English)
dataset = load_dataset("PolyAI/minds14", name="en-US", split="train")

# Select the first audio sample from the dataset
audio_input = dataset[0:4]["audio"][0]  # Get the first audio sample - se puede cambiar para ver diferente audios

# Access the audio file path (if available)
audio_path = audio_input["path"]  # Path to the audio file

# Play the selected audio (In Jupyter, this will play the audio directly)
Audio(audio_path)  # Plays the audio file from the path
```

**ASR** es una tecnología que permite a las computadoras o dispositivos convertir audio hablado en texto escrito. Se utiliza en aplicaciones como asistentes virtuales (como Alexa, Siri o Google Assistant), transcripción automática de reuniones, subtítulos automáticos, y muchas más.

**Wav2Vec2** es un modelo de reconocimiento de voz automático (ASR) desarrollado por Facebook AI Research. Es una versión avanzada del modelo Wav2Vec original y está diseñado para convertir directamente la onda sonora (audio) en vectores que luego se pueden usar para realizar tareas como transcripción de voz a texto.

Wav2Vec2 es un modelo basado en redes neuronales profundas, específicamente una red neuronal convolucional y transformers, y se entrenó utilizando aprendizaje no supervisado, lo que significa que fue entrenado con grandes cantidades de datos de audio sin etiquetas para aprender representaciones útiles del audio. Luego, se fine-tunea (ajusta) con datos etiquetados (audio con transcripciones) para mejorar el rendimiento en tareas de reconocimiento de voz.

```python
from transformers import pipeline
from datasets import load_dataset, Audio

asr = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h")

dataset = load_dataset("PolyAI/minds14", name="en-US", split="train")

dataset = dataset.cast_column("audio", Audio(sampling_rate=asr.feature_extractor.sampling_rate))

audio_inputs = dataset[:4]["audio"]
texts = asr(audio_inputs)
print([t["text"] for t in texts])
```
